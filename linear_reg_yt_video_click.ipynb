**Uji Multikolinearitas**

```python
variables = df[['Comments added', 'Shares', 'Likes', 'Average percentage viewed (%)', 'Views', 'RPM (USD)', 'Watch time (hours)']]
vifs = pd.Series(np.linalg.inv(variables.corr().to_numpy()).diagonal(), 
                 index=variables.columns, 
                 name='VIF')

vifs
```

**Menentukan Train, Test, dan Membuat Model Regresi**

```python
data_2.columns
```

```python
inputs = data_2.drop(['Views', 'Watch time (hours)', 'Impressions', 'Impressions click-through rate (%)'], axis=1)
X = inputs.copy()
y = X.pop('log_click')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# membuat model regresi
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# comment, like, views, impressions
reg.score(X_train, y_train)
```

```python
#comment, views, impressions
reg.score(X_train, y_train)
```

```python
# shares, view, impressions
reg.score(X_train, y_train)
```

```python
# data < 0.98 percentil
y_pred = reg.predict(X_test)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Target')
plt.ylabel('Prediction')
plt.show()
```

```python
# data < 0.99 percentil
y_pred = reg.predict(X_test)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Target')
plt.ylabel('Prediction')
plt.show()
```

```python
# membuat tabel prediksi
df_pred = pd.DataFrame(y_pred, columns=['Predictions']) 

df_pred.head()
```

```python
# mereset indeks dari y_test
y_test = y_test.reset_index(drop=True)
y_test
```

```python
df_pred['Target'] = y_test
df_pred
```

```python
# memeriksa residu/selisih/perbedaan nilai antara target dan prediksi
df_pred['Residual'] = df_pred['Target'] - df_pred['Predictions']
df_pred
```

```python
# persentase selisih, data < 0.98 percentil
df_pred['Difference%'] = (np.absolute(df_pred['Residual'])/(df_pred['Target'])*100)
df_pred.describe()
```
