**Uji Multikolinearitas**

kita periksa bagaimana korelasi antar variabel bebas
dengan menggunakan nilai VIF (Variance Inflation Factors)

```python
variables = data_cln[['Shares', 'Likes', 'Views', 'Watch time (hours)', 'Subscribers', 'Impressions']]
vifs = pd.Series(np.linalg.inv(variables.corr().to_numpy()).diagonal(), 
                 index=variables.columns, 
                 name='VIF')

vifs
```

dari hasil perhitungan VIF, variabel **Likes** dan **Views** memiliki nilai VIF yang tinggi (korelasi tinggi terhadap varabel bebas lainnya)  
sehingga varibael **Likes** dan **Views** tidak akan diikutsertakan dalam prediksi

```python
sns.pairplot(data_used, x_vars=['log_shares', 'log_watch_h', 'log_subscribers' ,'log_impressions'], y_vars='log_revenue',
             height=5, aspect=.9, kind='reg', diag_kind=None)
```

```python
variables = data_used[['Shares', 'Impressions', 'Subscribers', 'Watch time (hours)']]
vifs = pd.Series(np.linalg.inv(variables.corr().to_numpy()).diagonal(), 
                 index=variables.columns, 
                 name='VIF')

vifs
```

**Menentukan Train, Test, dan Membuat Model Regresi**

```python
data_used.columns
```

```python
inputs = data_used.drop(['Shares', 'Impressions', 'Subscribers', 'Watch time (hours)', 'Your estimated revenue (USD)'], axis=1)
X = inputs.copy()
y = X.pop('log_revenue')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# membuat model regresi
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
y_hat = reg.predict(X_train)
```

```python
# Plot residu/error
sns.distplot(y_train-y_hat)
plt.title('Residuals', size=18)

# Dari asumsi regresi terhadap zero mean, kita tahu bahwa error yang dihasilkan harus terdistribusi normal dengan mean 0
```

```python
reg_summary = pd.DataFrame(X.columns.values, columns=['Features'])
reg_summary['weights'] = reg.coef_
reg_summary
```

hasil yang didapatkan adalah:  
**Watch time (hour)** memiliki korelasi positif yang tinggi terhadap **Revenue**  
**Shares, Impression, Subscribers** juga memiliki korelasi positif, namun pengaruhnya kurang signifikan terhadap **Revenue**

**Testing**

```python
X_test.shape
```

```python
y_pred = reg.predict(X_test)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Target')
plt.ylabel('Prediction')
plt.show()
```

```python
# membuat tabel prediksi
df_pred = pd.DataFrame(y_pred, columns=['Predictions']) 

df_pred.head()
```

```python
# mereset indeks dari y_test
y_test = y_test.reset_index(drop=True)
y_test
```

```python
df_pred['Target'] = y_test
df_pred
```

```python
# memeriksa residu/selisih/perbedaan nilai antara target dan prediksi
df_pred['Residual'] = df_pred['Target'] - df_pred['Predictions']
df_pred
```

```python
# persentase selisih
df_pred['Difference%'] = (np.absolute(df_pred['Residual'])/(df_pred['Target'])*100)
df_pred.describe()
```
